```python
# Load the dataset
import pandas as pd
import numpy as np

df = pd.read_csv(r"K:\End-to-End-Airbnb-Price-Prediction-ML-System\artifacts\transformed_data\train_transformed.csv")
df.head()
```

```python
# Sperate the target variable and predictors
X = df.drop('log_price', axis=1)
y = df['log_price']
```

```python
# Split the data in the training and testing 
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20, random_state=42)

print(f"The X_train shape is {X_train.shape}")
print(f"The X_test shape is {X_test.shape}")
print(f"The y_train shape is {y_train.shape}")
print(f"The y_test shape is {y_test.shape}")
```

```python
# Import all requeried models and matrix
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
m1 = LinearRegression()
m1.fit(X_train, y_train)

m2 = DecisionTreeRegressor(random_state = 42)
m2.fit(X_train, y_train)

m3 = RandomForestRegressor(random_state = 42)
m3.fit(X_train, y_train)
```

```python
# Evaluation Matrix function
def evaluate_model(model, X_test, y_test, model_name):
  preds = model.predict(X_test)

  mae = mean_absolute_error(y_test, preds)
  rmse = np.sqrt(mean_squared_error(y_test, preds))
  r2 = r2_score(y_test, preds) * 100

  print(f"\n {model_name} Performace")
  print(f"MAE : {mae}")
  print(f"RMSE : {rmse}")
  print(f"R2 : {r2}")
```

```python
evaluate_model(m1, X_test, y_test, "Linear Regression")
evaluate_model(m2, X_test, y_test, "Decision Tree Regressor")
evaluate_model(m3, X_test, y_test, "Random Forest Regressor")
```

```python
# Try Hyper Parameters for Random Forest
rf = RandomForestRegressor(n_estimators=400,
                           max_depth=30,
                           min_samples_split=5,
                           min_samples_leaf=2,
                           n_jobs=-1,
                           random_state=42)
rf.fit(X_train, y_train)

evaluate_model(rf, X_test, y_test, "Random Forest Regressor")
```

```python
from xgboost import XGBRegressor

xgb = XGBRegressor(
    n_estimators=800,
    learning_rate=0.03,
    max_depth=10,
    subsample=0.8,
    colsample_bytree=0.8,
    random_state=42
)

xgb.fit(X_train, y_train)

evaluate_model(xgb, X_test, y_test, "XGBoost Tuned")
```
