# Handle Categorical Variables - One-Hot Encoding and Label Encoder

```python
# Load the dataset
import pandas as pd

df = pd.read_csv(r"K:\End-to-End-Airbnb-Price-Prediction-ML-System\data\raw\cleaned_airbnb_data.csv")
df.head()
```

```python
# Categorical columns
df.select_dtypes(include='object').columns
```

```python
# Check the how many unique property type we have in our dataset
df["property_type"].nunique()
```

```python
# Handels Property Type column by keeping only top 10 types and remaing will be other category
df['property_type'].value_counts().head(11)
```

```python
# Top 10 Property Types
top_property_types = df['property_type'].value_counts().nlargest(11).index

# Replace the Remaining Categories with "Other"
df['property_type'] = df['property_type'].apply(
    lambda x: x if x in top_property_types else 'Other'
)

# Check the Result
df['property_type'].value_counts()
```

```python
# Check the how many unique Cities we have in our dataset
df["city"].nunique()
```

```python
# Convert categorical columns to numerical 
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

df['room_type'] = le.fit_transform(df['room_type'])
df['bed_type'] = le.fit_transform(df['bed_type'])
df['cancellation_policy'] = le.fit_transform(df['cancellation_policy'])

df[["room_type","bed_type","cancellation_policy"]].head(10)
```

```python
#One Hot Encode Categorical Columns (by get_dummies method)
categorical_cols = df.select_dtypes(include='object').columns

df = pd.get_dummies(df, columns=categorical_cols, drop_first=False, dtype=int)

df.head(10)
```

```python
df.to_csv(r"K:\End-to-End-Airbnb-Price-Prediction-ML-System\data\raw\train.csv",index = False)
```
